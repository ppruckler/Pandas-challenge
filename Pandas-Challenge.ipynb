```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
itemCount=len(purchase_data["Item ID"].unique())
totalRev=purchase_data["Price"].sum()
averageP=totalRev/len(purchase_data["Item Name"])
purchaseNum=len(purchase_data["Item Name"])



item_df = pd.DataFrame({
    "Unique Items": [itemCount],
    "Average Price": [averageP],
    "Number of Purchases": [purchaseNum], 
    "Total Revenue": [totalRev]
})
item_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase = purchase_data[["Price", "Gender"]]
purchase.head()
test=purchase.groupby(["Gender"])
ave=test.mean()['Price']
su=test.sum()['Price']

banana = su/len(purchase_data["SN"])

purch2 = pd.DataFrame({"purchase count" : test.size(),
                       "average purchase price":ave, 
                       "total purchase value":su, 
                       "average total per person": banana
                      })



purch2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
test_test = purchase_data.loc[:,["SN", "Gender", "Age"]].drop_duplicates() 

bins = [0, 10,14,19,24,29,34,39, 500]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

test_test["age groups"] = pd.cut(test_test["Age"], bins, labels = group_names)

percent=(test_test["age groups"].value_counts()/len(test_test["SN"]))*100
age_df = pd.DataFrame({
    
    "Total Count": test_test["age groups"].value_counts(),
    "Percentage of Players": percent
})
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python


bins2 = [0, 10,14,19,24,29,34,39, 500]
group_names2 = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["age groups"] = pd.cut(purchase_data["Age"], bins2, labels = group_names2)

total_price = purchase_data.groupby(purchase_data["age groups"]).sum()["Price"]
total_purchase_value=(total_price/purchase_data["age groups"].value_counts())*purchase_data["age groups"].value_counts()

purch_df = pd.DataFrame({   
    "purchase count": purchase_data["age groups"].value_counts(),
    "average purchase Price": total_price/purchase_data["age groups"].value_counts(),
    "total purchase value":total_purchase_value,
    "average total purchase per person":total_price/purchase_data["age groups"].value_counts()/len(purchase_data)*1000
})
purch_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python


total_price3 = purchase_data.groupby(purchase_data["SN"]).sum()["Price"]


p_count = purchase_data["SN"].value_counts()

purch2_df = pd.DataFrame({   
    "purchase count": p_count,
    "average purchase price":total_price3/p_count,
    "total purchase value":p_count*(total_price3/p_count)
})
purch2_df.sort_values('total purchase value',ascending=False).head()
```
